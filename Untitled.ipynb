# Customer Churn Predection

## Problem Statement

Customer churn is a crucial concern for businesses across sectors. Understanding customer behaviors,
identifying key factors contributing to churn, and predicting when customers are likely to churn are vital
for reducing revenue loss and enhancing customer retention strategies.

In this project we will be building a machine learning model to predict customers' churn for a telecommunications company 

Dataset Overview:
The dataset contains the following columns:<br>
1- `customerID`: Customer ID<br>
2- `gender`: Customer's gender<br>
3- `SeniorCitizen`: Whether the customer is a senior citizen (1 for yes, 0 for no)<br>
4- `Partner`: Whether the customer has a partner<br>
5- `Dependents`: Whether the customer has dependents<br>
6- `tenure`: Number of months the customer has stayed with the company<br>
7- `PhoneService`: Whether the customer has phone service<br>
8- `MultipleLines`: Whether the customer has multiple phone lines<br>
9- `InternetService`: Type of internet service<br>
10- `OnlineSecurity`: Whether the customer has online security<br>
11- `OnlineBackup`: Whether the customer has online backup<br>
12- `DeviceProtection`: Whether the customer has device protection<br>
13- `TechSupport`: Whether the customer has tech support<br>
14- `StreamingTV`: Whether the customer streams TV<br>
15- `StreamingMovies`: Whether the customer streams movies<br>
16- `Contract`: Type of contract (e.g., month-to-month, one year, two years)<br>
17- `PaperlessBilling`: Whether the customer uses paperless billing<br>
18- `PaymentMethod`: Payment method (e.g., electronic check, mailed check)<br>
19- `MonthlyCharges`: Monthly charges<br>
20- `TotalCharges`: Total charges<br>
21- `Churn`: Target variable, indicating whether the customer churned (1 for yes, 0 for no)

## Feature Engineering & Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
df = df.drop('customerID', axis=1)
```

### Label Encoding the Features

```python
columns = list(df.columns)
columns.remove('tenure')
columns.remove('MonthlyCharges')
columns.remove('TotalCharges')

label_encoded_models = []

for i in columns:
    le = LabelEncoder()
    le.fit(df[i])
    df[i] = le.transform(df[i])
    label_encoded_models.append(le)
label_encoded_models = label_encoded_models[:-1]
df.head()
```

```python
sns.heatmap(df.corr())
```

### Feature Selection

```python
x = df.drop('Churn', axis=1)
y = df['Churn']

x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
selector = SelectKBest(chi2, k=14)
x_train = selector.fit_transform(x_train, y_train)
x_test = selector.transform(x_test)
```

#### these are the features that were selected

```python
list(selector.get_feature_names_out())
```

## Machine Learning Model Development

```python
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import lightgbm as lgb
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier, GradientBoostingClassifier,\
                             HistGradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix,\
                            precision_score, recall_score, ConfusionMatrixDisplay
```

```python
model_dict = {
    'RandomForestClassifier': RandomForestClassifier(),
    'LogisticRegression':LogisticRegression(solver='liblinear'),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'BaggingClassifier': BaggingClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'ExtraTreesClassifier' : ExtraTreesClassifier(),
    'GradientBoostingClassifier' : GradientBoostingClassifier(),
    'HistGradientBoostingClassifier': HistGradientBoostingClassifier(),
    'XGBClassifier': XGBClassifier(),
}
```

```python
model_name = []

model_cv = []
model_roc = []
model_accuracy = []
model_f1_score = []
model_precision_score = []
model_recall_score = []

model_ypred = []

kfold = KFold(n_splits=10, shuffle=True)

for i in model_dict:
    model = model_dict[i]
    scores = cross_val_score(model, x_train, y_train, cv=kfold)    
    model.fit(x_train,y_train)
    
    model_name.append(i)
    model_cv.append(np.mean(scores))
    
    y_pred = model.predict(x_test)
    model_ypred.append(y_pred)
    
    model_roc.append(roc_auc_score(y_test,y_pred))
    model_accuracy.append(accuracy_score(y_test,y_pred))
    model_f1_score.append(f1_score(y_test,y_pred))
    model_precision_score.append(precision_score(y_test,y_pred))
    model_recall_score.append(recall_score(y_test,y_pred))
```

## Model Evaluation

```python
dict_model_info = {
    'Cross Validation Score':model_cv,
    'ROC AUC':model_roc,
    'Accuracy Score':model_accuracy,
    'F1 Score':model_f1_score,
    'Precision Score':model_precision_score,
    'Recall Score':model_recall_score
}

df_model_eva = pd.DataFrame(dict_model_info, index=model_name)
df_model_eva.sort_values(['Cross Validation Score','ROC AUC','Accuracy Score','F1 Score','Precision Score','Recall Score'],
                         axis=0, ascending=False, inplace=True)
df_model_eva
```

### Confusion Matrix

```python
fig, axes = plt.subplots(3,3, figsize = (12,12))

index1=0
index2=0
for i in model_dict:
    if index1>2:
        index2 += 1
        index1 = 0 
    model=model_dict[i]
    ConfusionMatrixDisplay(confusion_matrix(y_test, model.predict(x_test))).plot(ax=axes[index1,index2])
    
    axes[index1,index2].set_title(i)
    index1+=1
plt.show()
```

## Recommendations

### EDA

#### Customer Segmentation Analysis

Understanding customer characteristics paints a picture of a customer with strong ties to their current service provider, suggesting a lower likelihood of churn. 

Factors like being male, young, and married, along with dependencies, imply a stable personal life and a reliance on existing services for communication and connectivity needs. Additionally, committing to two-year contracts and opting for traditional paper billing indicates a preference for long-term relationships and resistance to change, further reducing the probability of churn.

Moreover, the customer's reliance on credit cards for payments not only signifies financial stability but also establishes a billing relationship with the service provider, adding another layer of inertia against switching. Collectively, these traits underscore the importance of understanding customer demographics, preferences, and behaviors in devising retention strategies. By recognizing and catering to the needs of such customers, service providers can foster stronger loyalty and mitigate churn, ultimately contributing to long-term business success.


 
By studying these unique customer characteristics, the company can tailor its advertising efforts to draw comparable individuals, to expand its client base within the marketplace.

#### Feature Selection

Based on the feature selection step, we have acquired the best features that will make a difference, and here are some actions that we could take to leverage these features.

1. **Senior Bundles:** Offer elderly-friendly plans with larger fonts, pre-set protection, and extended tech assist Offer simplified plans with clear causes, prioritize customer service help, and explore options for bundled offerings with internet security and tool protection. This would be a great step since elderly customers are more likely to churn.
2. **Family Upsell:** Recommend parental controls, more statistics plans, multi-line discounts for families, and explore offering entertainment programs suitable for families. This would be great since customers with children are less likely to churn.
3. **Loyalty Rewards:** After analyzing tenure, we can see that long-term customers are less likely to churn, we can keep this up by providing attractive contracts with increasing benefits for unswerving customers, and emphasizing flexibility for no-agreement clients.
4. **Paperless billing & payment methods:** Promote paperless billing and automatic payment facility to reduce churn. We can offer a wider range of payment methods to accommodate different preferences to ensure seamless integration and user-friendly interfaces for these payment methods to enhance customer experience.
5. **Monthly Costs:** Introducing cost-effective plans or packages tailored to customers with high monthly expenses would lower churn rates. These plans should provide value for money and align with the specific needs and preferences of this customer segment. we can highlight the cost savings and benefits of these plans to attract and retain price-sensitive customers.

### The Best Model

#### Logistic Regression V.S. AdaBoostClassifier V.S. GradientBoostClassifier

The model training step gave some interesting insights, after training multiple models, these were the top 3 performing models, these models gave an extremely close result with the measuring metrics, however, after a lot of testing, it determined that the Logistic Regression is the best fitting model for the problem.

The task was to predict whether a customer would churn or not, so it's considered a binary classification problem. using a Booster Classifier to do this task would increase the cost, and take time.

While AdaBoost and GradientBoost might offer slightly higher accuracy in some cases, the trade-off in interpretability for this specific scenario might not be worth it. so sticking with Logistic Regression is the best choice for this project.
